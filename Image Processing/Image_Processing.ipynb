```python
# WebCam - Flip and Crop

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

cam = cv.VideoCapture(0)
while True:
    _, img = cam.read()
    cv.imshow('Original', img)
    
    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]
    cv.imshow('Frame0', r)
    cv.imshow('Frame1', g)
    cv.imshow('Frame2', b)
    
    flipped = cv.flip(img,1)
    cropped = img[500:800, 800:1200, :]
    cv.imshow('Flipped', flipped)
    cv.imshow('Cropped', cropped)

    
    hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
    h = hsv[:,:,0]
    s = hsv[:,:,1]
    v = hsv[:,:,2]
    cv.imshow('HSV', hsv)
    cv.imshow('Hue - POPupColor', h)
    cv.imshow('Saturation - BrightColor', s)
    cv.imshow('Value - DarkColor', v)
    
    gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
    cv.imshow('Gray', gray)

    key = cv.waitKey(20)
    
    if (key == 27):
        cam.release()
        break
    elif (key == 13):
        cv.imwrite('Selfie.png', img)
        cam.release()
        break
```

### Increasing Brightness of WebCam

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# img + 10 = adding 10 to each pixel color in overall pic

pixels = float(10)
cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    img = cv.flip(img, 1)

    bright_1 = (img + pixels)
    bright_1[bright_1 > 255] = 255
    bright_1 = bright_1.astype(np.uint8)
                               
    bright_2 = (img + 2*pixels)
    bright_2[bright_2 > 255] = 255
    bright_2 = bright_2.astype(np.uint8)
    
    bright_3 = (img + 3*pixels)
    bright_3[bright_3 > 255] = 255
    bright_3 = bright_3.astype(np.uint8)

    cv.imshow('original', img)
    cv.imshow('Bright_1', bright_1)
    cv.imshow('Bright_2', bright_2)
    cv.imshow('Bright_3', bright_3)

    key = cv.waitKey(10)
    if key == 27:
        cam.release()
        cv.destroyAllWindows()
        cv.waitKey(2000)
    
```

### Reducing Brightness of Webcam

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# img + 10 = adding 10 to each pixel color in overall pic

pixels = float(10)
cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    img = cv.flip(img, 1)

    bright_1 = (img - pixels)
    bright_1[bright_1 < 0] = 0
    bright_1 = bright_1.astype(np.uint8)
                               
    bright_2 = (img - 2*pixels)
    bright_2[bright_2 < 0] = 0
    bright_2 = bright_2.astype(np.uint8)
    
    bright_3 = (img - 3*pixels)
    bright_3[bright_3 < 0] = 0
    bright_3 = bright_3.astype(np.uint8)

    cv.imshow('original', img)
    cv.imshow('Bright_1', bright_1)
    cv.imshow('Bright_2', bright_2)
    cv.imshow('Bright_3', bright_3)

    key = cv.waitKey(10)
    if key == 27:
        cam.release()
        cv.destroyAllWindows()
        cv.waitKey(2000)
    
```

### Warm Tone to Images

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def add_tone(img_path, color, a, b):
    
    background = []

    img = cv.imread(img_path)

    cols = img.shape[1]
    rows = img.shape[0]


    for i in range(rows):
        temp = []
        for j in range(cols):
            temp.append(color)

        background.append(temp)

    background = np.array(background).astype(np.uint8)

    final = cv.addWeighted(img, a , background, b , 0)

    cv.imshow('Original',img)
    cv.waitKey(0)

    cv.imshow('Processed',final)
    cv.waitKey(0)
```

### Setting aspect ratio to auto fit logo in various place

```python

# 1. Top Right
# Height (1% = Margin | 7% = Logo | 92% = Vacant) 720

# Width (95.5% = Margin | 4% = Logo | .5% = Vacant) 1280

import cv2 as cv
import numpy as np
 
cam = cv.VideoCapture(0)

logo = cv.imread('Terapanth_Official_Symbol.png')

while True:
    
    _, img = cam.read()

    img = cv.flip(img,1)
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    margin_top  = int(ht/100)             # 1%
    logo_height = int((ht/100)*7)         # 7%
    
    margin_left = int((wt/100)*95.5)      # 95.5%
    logo_width  = int((wt/100)*4)         # 4%
    
    
    logo = cv.resize(logo, (logo_width, logo_height))
    
    
    img[ margin_top : logo_height + margin_top , margin_left : margin_left + logo_width ] = logo

    cv.imshow('Frame'  , img)
     
    if cv.waitKey(1) == 27:
        cam.release()
        break
```

```python
import cv2 as cv
import numpy as np
classifier = cv.CascadeClassifier('haarcascade_frontalface_default.xml')

cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    img = cv.flip(img,1)
    
    faces = classifier.detectMultiScale(img, 1.1, 5)
    
    for f in faces:                             # Going through Each Face Detected
        if f[-1] == max(faces[:,-1]):           # Finding the Face with Maximum Area
            break 

    if (len(faces) >= 1):                       # Drawing Rectange on the Face
        x = f[0] 
        y = f[1] 
        w = f[2]
        h = f[3]

        cv.rectangle(img, (x,y),(x+w,y+h) , (0,180,0), 4)   
        face = img[y:y+h, x:x+w]
            
        face = cv.resize(face, (256,256))

    cv.imshow('Frame'  , img )
    cv.imshow('Face'   , face)
    
    if cv.waitKey(1) == 27:
        cam.release()
        break
 
```

```python

import os
import cv2 as cv
import numpy as np


img = cv.imread('Terapanth_Gurus.png')
classifier = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
faces = classifier.detectMultiScale(img, 1.1, 5)


def save(frame, folder_name): 
    name_img = len(os.listdir(folder_name)) + 1
    name_img = folder_name + "/IMG_" + str(name_img) + '.png'
    cv.imwrite(name_img, frame)
    print(name_img ,'is exported')

for (x,y,w,h) in faces:
    face = img[y:y+h, x:x+w]
    cv.imshow('Photo',img)
    cv.imshow('Face'   , face)
    
    if cv.waitKey(0) == 13:         # Save the Image | 13  = Enter Key
        save(face, 'Terapanth')
    
    elif cv.waitKey(0) == 127:      # Skip the Image | 127 = Delete Key
        pass
```
